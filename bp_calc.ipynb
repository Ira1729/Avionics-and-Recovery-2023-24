```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
psi = np.arange(10,16,1)
pressure = psi*6894.76

r_inner = 0.015
R = 8.314
T = 500
MM = 867.7 # in gms
piston_mass = 0.9

height=[3,2.5]
c = 0.05
t = 0.001

```

```python
for h in height:
    w = (h*(MM*pressure*np.pi*(r_inner**2))*np.exp(-c*t/piston_mass))/(R*T*100)
    print(psi,w)
    plt.plot(psi,w,'.-',label=f'{h}')
    
plt.xlabel('Pressure in Psi')
plt.ylabel('quantity of black powder in gms')
plt.grid()
plt.legend()
plt.show()
```
